## 1.1. Još o SVD faktorizaciji  i primjenama


```python
from PIL import Image
slika = Image.open('macka1.jpg')
plt.figure(figsize=(5, 4))
plt.imshow(slika);
```

```python
matrica = np.array(list(slika.getdata(band=0)), float)
matrica.shape = (slika.size[1], slika.size[0])
matrica = np.matrix(matrica)
plt.figure(figsize=(5,4))
plt.imshow(matrica, cmap='gray');
```

```python
U, sigma, V = np.linalg.svd(matrica)

for i in range(5, 36, 10):
    manji_rang = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(manji_rang, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

U ovom dijelu predavanja reći ćemo ukratko nešto o primjenama SVD dekompozicije matrice. Sljedeće predavanje morate znati ugrubo, dakle otprilike objasniti kako nam koristi SVD, a ne morate promatrati detalje.
